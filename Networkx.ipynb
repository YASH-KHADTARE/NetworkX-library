```python
#basics
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#adding vetex , edges
#adding vertices(nodes)
G=nx.Graph()
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(7)
G.add_node(9)
#adding edges
G.add_edge(1,2)
G.add_edge(3,1)
G.add_edge(2,4)
G.add_edge(4,1)
G.add_edge(9,1)
G.add_edge(1,7)
G.add_edge(2,9)

 
# To find all the neighbor of a node
neighbor_list = G.neighbors(2)
print("neighbor_list = ",neighbor_list)

nx.draw(G, with_labels = True)
#nx.draw_planar(G, with_labels = True)
plt.show()

```

```python
# To get all the nodes of a graph
node_list = G.nodes()
print("node list = ",node_list)
  
# To get all the edges of a graph
edge_list = G.edges()
print("edge list = ",edge_list)

```

```python
# To remove a node of a graph
G.remove_node(3)
node_list = G.nodes()
print("node_list = ",node_list)

  
# To remove an edge of a graph
G.remove_edge(1,2)
edge_list = G.edges()
print("edge_list = ",edge_list)

nx.draw(G, with_labels = True)
plt.show()

```

```python
# To find number of nodes
n = G.number_of_nodes()
print("nodes =",  n)

  
# To find number of edges
m = G.number_of_edges()
print("edges= ",m)

  
# To find degree of a node
# d will store degree of node 2
d = G.degree(2)
print("degree of vertex 2 = ",d)
```

```python
#operations on graph
G=nx.Graph()

G.add_nodes_from([1,2,3,4])
G.add_edges_from([(1,2),(2,3),(3,4),(4,1)])
nx.draw(G, with_labels=True)
```

```python
H=nx.Graph()
H.add_nodes_from([3,4,5,6])
H.add_edges_from([(3,4),(4,5),(5,6),(6,3)])
nx.draw(H,with_labels=True)
```

```python
I=nx.intersection(G,H)
complement=nx.complement(G)
```

```python
nx.draw(I,with_labels=True)
```

```python
nx.draw(complement,with_labels=True)
```

```python
#Directed graph
G=nx.DiGraph()

G.add_nodes_from([1,2,3,4,5])
G.add_edges_from([(1,2),(2,3),(3,4),(4,1),(4,5)])
nx.draw(G,with_labels=True)
```

```python
G=nx.DiGraph()
G.add_nodes_from([1,2,3,4,5,6,7,8])
G.add_edges_from([(1,2),(2,3),(3,4),(4,1),(4,5),(6,5),(6,7),(7,8),(8,6)])
nx.draw_planar(G,with_labels=True)
```

```python
nx.strongly_connected_components(G)
```

```python
for i in nx.strongly_connected_components(G):
  print(i)
```
